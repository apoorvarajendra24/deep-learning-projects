```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = [
    "The cat jumped.",
    "Rain started suddenly.",
    "He dropped his keys.",
    "The light flickered.",
    "She hummed softly.",
    "A dog barked nearby.",
    "The phone buzzed.",
    "Wind shook the door.",
    "The cup fell.",
    "Birds flew away."
]
sentences
```

```python
# Defining the vocab size
vocab_size = 1000
```

```python
# One hot representation
one_hot_representation = []
for words in sentences:
    one_hot_representation.append(one_hot(words,vocab_size))

one_hot_representation
```

```python
# Word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences   # to make the sequences of same size
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
# Set a maximum length of padding for sentences
sent_length = 7
embedded_docs = pad_sequences(one_hot_representation,padding='pre',maxlen=sent_length)
embedded_docs
```

```python
# We need to convert each word of the sentence into a seperate vector
# Feature representatiom
dimensions = 10  # number of feature representaion you use to convert your vocab words into vectors

model = Sequential()
model.add(Embedding(vocab_size, dimensions, input_length=sent_length))
model.compile(optimizer='adam',loss='mse')
```

```python
model.predict(embedded_docs)
```
